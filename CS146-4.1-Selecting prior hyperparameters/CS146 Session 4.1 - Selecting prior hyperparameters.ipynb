```python
from scipy.optimize import minimize
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-(x-2)**2)

def minus_f(x):
    #when maximization is required, minimize the negative of the function to be maximized
    return -f(x)

result = minimize(minus_f,1)
x_opt = result.x
print(x_opt)

x = np.linspace(0,4)
plt.plot(x,f(x))
plt.show()
```

```python
def f(x):
    return np.sin(x * np.cos(x))

plot_x = np.linspace(0, 10, 500)
plot_y = f(plot_x)
plt.plot(plot_x, plot_y)

for x_initial in [2, 4, 6, 8]:
    result = minimize(f, x_initial)
    x_final = result.x
    plt.plot([x_initial, x_final], [f(x_initial), f(x_final)], 'o')
    print('Started at %.2f; ended at %.2f; f(%.2f) = %.2f' % (
    x_initial, x_final, x_final, f(x_final)))
    
plt.show()
```

Normal-inv-gamma

<b>Mean</b>

$E(x)=\mu$

$E(\sigma^2)=\frac{\beta}{\alpha-1}$

<b>Variance</b>

$Var(x)=\frac{\beta}{(\alpha-1)\lambda}$

$Var(\sigma^2)=\frac{\beta^2}{(\alpha-1)^2(\alpha-2)}$

```python
#mu = 2.3
#beta/(alpha-1) = 2.75
#beta/((alpha-1)*lambda_) = 0.5^2
#beta^2/((alpha-1)^2*(alpha-2)) = 1^2
#x[0] = mu
#x[1] = lambda
#x[2] = alpha
#x[3] = beta

def f(x):
    return (x[0]-2.3)**2 + ((x[3]/(x[2]-1))-2.75)**2 + ((x[3]/((x[2]-1)*x[1]))-0.5**2)**2 + ((x[3]**2/((x[2]-1)**2*(x[2]-2)))-1)**2

for x_initial in [
np.array([2,0.5,8,8]),
np.array([2,3,7,1]),
np.array([1,2,3,1]),
np.array([4,5,3,2])
]:
    result = minimize(f, x_initial)
    x_final = result.x
    print('Started at', x_initial)
    print('Ended at', x_final)
    print('f(%s) = %.6f' % (x_final, f(x_final)))
    print()
    
```

```python
from scipy import stats
import seaborn as sns
sns.set()

def norminvgamma_rvs(mu, nu, alpha, beta, size=1):
    '''
    Generate n samples from the normal-inverse-gamma distribution. This function
    returns a (size x 2) matrix where each row contains a sample, (x, sigma2).
    '''
    sigma2 = stats.invgamma.rvs(a=alpha, scale=beta, size=size)  # Sample sigma^2 from the inverse-gamma
    x = stats.norm.rvs(loc=mu, scale=np.sqrt(sigma2 / nu), size=size)  # Sample x from the normal
    return np.vstack((x, sigma2)).transpose()

plt.figure(figsize=(12, 6))
x = np.linspace(0,5,1000)
samples = norminvgamma_rvs(2.30000008, 11.00143402,  9.56252627, 23.5470067, size=10000)
    
print("68% confidence interval for mean:", (np.percentile(samples[:,0], 16), np.percentile(samples[:,0], 84)))
print("68% confidence interval for variance:", (np.percentile(samples[:,1], 16), np.percentile(samples[:,1], 84)))
```
