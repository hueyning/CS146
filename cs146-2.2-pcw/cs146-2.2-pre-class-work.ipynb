```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import seaborn as sns
sns.set()

alpha = 1
beta = 2
gamma_dist1 = stats.gamma(a=alpha, scale=1/beta)

alpha = 2
beta = 1
gamma_dist2 = stats.gamma(a=alpha, scale=1/beta)

plt.figure(figsize=(8, 4))
x = np.linspace(0, 5)
plt.plot(x, gamma_dist1.pdf(x), color='#FA746D' )
plt.plot(x, gamma_dist2.pdf(x), color='#6DAFFA' )
plt.title('Gamma probability density function')
pink = mlines.Line2D([], [], color='#FA746D', marker='',
                          markersize=15, label='$alpha=1,beta=2$')
blue = mlines.Line2D([], [], color='#6DAFFA', marker='',
                          markersize=15, label='$alpha=2,beta=1$')
plt.legend(handles=[pink,blue])
```

```python
alpha = 1
beta = 2
gamma_dist = stats.gamma(a=alpha, scale=1/beta)
sample_gamma = np.random.gamma(alpha,beta,100)

plt.figure(figsize=(8, 4))
plt.hist(sample_gamma,50,density=True)
x = np.linspace(0, 5)
plt.plot(x, gamma_dist.pdf(x))
```

```python
mu=0
sigma=1
norm_dist1 = stats.norm(loc=mu, scale=sigma)

mu=-2
sigma=4
norm_dist2 = stats.norm(loc=mu, scale=sigma)

plt.figure(figsize=(8, 4))
x = np.linspace(-15, 10, 1000)
plt.plot(x, norm_dist1.pdf(x), color='#FA746D' )
plt.plot(x, norm_dist2.pdf(x), color='#6DAFFA' )
plt.title('Normal probability density function')
pink = mlines.Line2D([], [], color='#FA746D', marker='',
                          markersize=15, label='$mu=0,sigma=1$')
blue = mlines.Line2D([], [], color='#6DAFFA', marker='',
                          markersize=15, label='$mu=-2,sigma=4$')
plt.legend(handles=[pink,blue])
```

```python
mu=0
sigma=1
norm_dist = stats.norm(loc=mu, scale=sigma)
sample_norm = np.random.normal(mu,sigma,100)

plt.figure(figsize=(8, 4))
plt.hist(sample_norm,50,density=True)
x = np.linspace(-5,5)
plt.plot(x, norm_dist.pdf(x))
```

```python
alpha = 2
beta = 2
beta_dist1 = stats.beta(a=alpha, b=beta)

alpha = 4
beta = 10
beta_dist2 = stats.beta(a=alpha, b=beta)

plt.figure(figsize=(8, 4))
x = np.linspace(0,5,1000)
plt.plot(x, beta_dist1.pdf(x), color='#FA746D' )
plt.plot(x, beta_dist2.pdf(x), color='#6DAFFA' )
plt.title('Beta probability density function')
pink = mlines.Line2D([], [], color='#FA746D', marker='',
                          markersize=15, label='$alpha=2,beta=2$')
blue = mlines.Line2D([], [], color='#6DAFFA', marker='',
                          markersize=15, label='$alpha=4,beta=10$')
plt.legend(handles=[pink,blue])
```

```python
alpha = 2
beta = 2
beta_dist = stats.beta(a=alpha, b=beta)
sample_beta = np.random.beta(alpha,beta,100)

plt.figure(figsize=(8, 4))
plt.hist(sample_beta,50,density=True)
x = np.linspace(-0.5, 1.5)
plt.plot(x, beta_dist.pdf(x))
```
